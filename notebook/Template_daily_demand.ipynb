```python
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00409/Daily_Demand_Forecasting_Orders.csv"
base_name = os.path.basename(data_url)
with timer("read"):
    file_path = os.path.join("../data/input", base_name)
    urllib.request.urlretrieve(data_url, file_path)
    df = pd.read_csv(file_path, index_col=None, encoding="utf-8", sep=";")
```
